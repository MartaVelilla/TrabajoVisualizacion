```python
import pandas as pd

df = pd.read_csv("dataexport_20251003T112456.csv", sep=",")

print(df.head())
```

```python
# quito las primeras 3 filas que no contienen datos relevantes
df = df.iloc[3:]
 
# renombro las columnas con fila 4
df.columns = df.iloc[0]

df = df[7:]

df
```

```python

# 2️⃣ Convertir timestamp a datetime
df["variable"] = pd.to_datetime(df["variable"], errors="coerce")

# 3️⃣ Eliminar filas sin timestamp válido
df = df.dropna(subset=["variable"])

# 4️⃣ Extraer la fecha (sin hora)
df["fecha"] = df["variable"].dt.date

# 5️⃣ Convertir todas las columnas posibles a numéricas (ignorando errores)
for col in df.columns:
    if col not in ["variable", "fecha"]:
        df[col] = pd.to_numeric(df[col], errors="coerce")

# 6️⃣ Calcular estadísticas diarias
df_diario = df.groupby("fecha").agg(['mean', 'min', 'max'])

# 7️⃣ Aplanar el multi-índice de columnas (para que los nombres sean simples)
df_diario.columns = [f"{col}_{stat}" for col, stat in df_diario.columns]

# 8️⃣ Resetear el índice para que 'fecha' sea una columna normal
df_diario = df_diario.reset_index()

# 9️⃣ Guardar el nuevo CSV con medias, mínimos y máximos diarios
#df_diario.to_csv("datos_diarios_completos.csv", index=False)

print("✅ Archivo 'datos_diarios_completos.csv' creado correctamente con medias, mínimos y máximos.")
print(df_diario.head())
```

```python
import pandas as pd

# Cargar el dataset original
df = pd.read_csv("datos_diarios_completos.csv")

# --- 1️⃣ Eliminar columnas vacías o sin información útil ---
df = df.drop(columns=["variable_mean", "variable_min", "variable_max"], errors="ignore")

# --- 2️⃣ Filtrar solo las columnas de interés ---
# Nos quedamos con la fecha y las columnas que terminan en "_mean"
cols_to_keep = ["fecha"] + [col for col in df.columns if col.endswith("_mean")]
df_limpio = df[cols_to_keep]

# --- 3️⃣ (Opcional) Renombrar columnas para mayor claridad ---
df_limpio = df_limpio.rename(columns=lambda x: x.replace("_mean", "").strip())

# --- 4️⃣ Guardar el nuevo dataset limpio ---
df_limpio.to_csv("datos_diarios_limpios.csv", index=False)

print("✅ Dataset limpio guardado como 'datos_diarios_limpios.csv'")
print("Columnas finales:")
print(df_limpio.columns.tolist())
```
